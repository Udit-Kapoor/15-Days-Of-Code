PROCEDURAL ORIENTED PROGRAMMING - TRADITIONAL PROGRAMMING TECHNIQUE

TOP DOWN APPROCH - FUNCTIONS ARE GIVEN MORE IMPORTANCE IN PROCEDURAL PROGRAMMING

```
def Add(a,b,c):
  return a+b+c

def Mul(a,b,c):
  return a*b*c

print("Addition : ",Add(5,7,8))
print("Multiplication : ",Mul(5,7,8))
```

OBJECT ORIENTED PROGRAMMING - BOTTOM UP APPROACH. fOLLOWS THE CRITERION THAT DATA IS GIVEN MORE IMPORTANCE

```
class A:
  my_var=10

  def Add(self,param):
    self.my_var=self.my_var+param.my_var
    return self.my_var

  def Mul(self,param):
    self.my_var=self.my_var*param.my_var
    return self.my_var

a1=A()
print(a1.my_var)
a1.my_var=7
a2=A()
a2.my_var=100
print("Addition :",a2.Add(a1))
print(a2.my_var)

a3=A()
a3.my_var=78
print("Addition : ",a2.Add(a3))
print(a2.my_var)

a4=A()
a4.my_var=78
print("Addition : ",a2.Add(a4))
print(a2.my_var)
 
```

INHERITANCE - THINGS, MONEY PROPERTY INHERITED FROM PARENTS OR GRANDPARENTS

##SINGLE INHERITANCE

###Class Student - name, roll no, admn no

##Class Marks - Inherit Student (phy,chem,maths)

```
class Student:

  def __init__(self,name='Mahima',roll=98,admn=567):
    self.name=name
    self.roll=roll
    self.admn=admn
  
  def input(self):
    self.name=input("Enter name : ")
    self.roll=int(input('Enter roll : '))
    self.admn=int(input('Enter admn no :'))

  def display(self):
    print("\nName :",self.name,"\tRoll No: ",self.roll,"\tAdmission No.",self.admn)

s1=Student()
s1.display()
s1.input()
s1.display()
```

```
class Marks(Student):

  def __init__(self):
    super().__init__()
    self.phy=0
    self.chem=0
    self.maths=0
  
  def input(self):
    super().input()
    self.phy=int(input("Enter marks in phy :"))
    self.chem=int(input("Enter marks in chem :"))
    self.maths=int(input("Enter marks in maths :"))
  
  def display(self):
    super().display()
    print("Physics :",self.phy,"\tChemistry: ",self.chem,"\tMaths:",self.maths)

m1=Marks()
m1.display()
m1.input()
m1.display()
```

MULTILEVEL INHERITANCE 

Grandfather->Father->You(Child)

```
class Average(Marks):
  def __init__(self):
    super().__init__()
    self.total=0

  def display(self):
    self.total=self.phy+self.chem+self.maths
    super().display()
    print("Total :",self.total)
    print("Average: ",self.total/3)

a1=Average()
a1.display()
a1.input()
a1.display()
```

MULTIPLE INHERITANCE - Inherit from both parents - Father and Mother (Parent)

Child Class - You

```
class Papa:
  def __init__(self,name="Ashok",money=10000):
    self.papa_name=name
    self.papa_money=money
  
  def display(self):
    print("Papa's name : ",self.papa_name)
    print("Papa's Money :",self.papa_money)

p1=Papa('Abhinav',17000)
p1.display()
```

```
class Mom:
  def __init__(self,name="Mamta",money=20000):
    self.mom_name=name
    self.mom_money=money
  
  def display(self):
    print("Mom's name : ",self.mom_name)
    print("Mom's Money :",self.mom_money)

m1=Mom()
m1.display()
```

```
class Child(Papa,Mom):
  def __init__(self,name="Mahima",money=2000):
    Papa.__init__(self)
    Mom.__init__(self)
    self.name=name
    self.money=money
  
  def display(self):
    Papa.display(self)
    Mom.display(self)
    print("My name : ",self.name)
    print("My Money :",self.money)

c1=Child()
c1.display()
```
