### Handling an exception


```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("division by zero! not allowed")
    else:
        print("result is", result)
    finally:
        print("finally clause - always executed")
```

```python
divide(2, 1)
```

```python
divide(2, 0)
```

```python
divide("2", "1")
```

### Multiple exceptions

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("division by zero!")
    except TypeError:
        print("unsupported operand type(s)")
        
    else:
        print("result is", result)
    finally:
        print("finally clause - always executed")
```

```python
divide("2", "1")
```

```python
def divide(x, y):
    try:
        result = x / y
    except(ZeroDivisionError, TypeError):
        print("Error Occurred and Handled")
    else:
        print("result is", result)
    finally:
        print("finally clause - always executed")
```

```python
divide("2", "1")
```

### Python facilitates us to not specify the exception with the except statement.

```python
def divide(x, y):
    try:
        result = x / y
    except:
        print("Something else went wrong")
    else:
        print("result is", result)
    finally:
        print("finally clause - always executed")
```

```python
divide(2, 0)
```
