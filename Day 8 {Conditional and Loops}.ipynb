# IF-ELSE

```python
x=5
if x>0:
    print("positive number")
else:
    print("negative number")
```

# IF-ELIF

```python
num = 0

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
```

# jump statement- break and continue

```python
num_sum = 0
count = 0
while(count<10):
    num_sum = num_sum + count
    count = count + 1 
    if count== 5:
        break
print("Sum of first ",count,"integers is: ", num_sum)
```

```python
for x in range(7):
    if (x == 3 or x==6):
        continue
    print(x)
```
